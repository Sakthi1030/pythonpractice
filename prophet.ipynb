```python
from prophet import Prophet
import pandas as pd  
#creating a datframe
data = {'ds' : pd.date_range(start = '2024-01-01',periods = 100),
       'y':(x + (x % 5) for x in range(100))}
df = pd.DataFrame(data)
#fiting data in the model 
model = Prophet()
model.fit(df)
#defining number of feature period to determine
future = model.make_future_dataframe(periods=30)
#predict
pred = model.predict(future)
#visualize the prediction
model.plot(pred)
```
